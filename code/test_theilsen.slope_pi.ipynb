```python
t_mean = [16, 20, 24]
t_p025 = [10,8,14]
t_p975 = [18,18,22]
time = [1,2,3]
```

```python
import numpy as np

def extended_theil_sen_slope(time, t_mean, t_p025, t_p975):
    # Prepare all combinations of temperature values
    temperatures = [t_mean, t_p025, t_p975]
    
    # Initialize list to store slopes
    slopes = []
    
    # Calculate slopes for each pair of time points
    for i in range(len(time)):
        for j in range(i + 1, len(time)):
            # Calculate slope for each combination
            for temp_i in temperatures:
                for temp_j in temperatures:
                    slope = (temp_j[j] - temp_i[i]) / (time[j] - time[i])
                    slopes.append(slope)
    
    # Return the median slope
    return np.mean(slopes)

# Example usage:
t_mean = [16, 20, 28]
t_p025 = [14, 12, 2]
t_p975 = [28, 26, 29]
time = [0,1,2]

extended_slope = extended_theil_sen_slope(time, t_mean, t_p025, t_p975)
print(f"The extended Theil-Sen slope is: {extended_slope}")
```

```python
import numpy as np
from scipy.stats import theilslopes

def standard_theil_sen_slope(time, t_mean):
    # Use scipy's theilslopes function
    slope, intercept, lower_slope, upper_slope = theilslopes(t_mean, time)
    return slope


standard_slope = standard_theil_sen_slope(time, t_mean)
print(f"The standard Theil-Sen slope is: {standard_slope}")
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(10, 6))

# Plot the filled area between t_p025 and t_p975
plt.fill_between(time, t_p025, t_p975, color='gray', alpha=0.3, label='Uncertainty Range (P025 to P975)')

# Plot the mean temperature with a line
plt.plot(time, t_mean, color='blue', label='Mean Temperature')

intercept = t_mean[0]
# Plot the standard Theil-Sen slope
plt.plot(time, intercept + standard_slope * np.array(time), color='red', linestyle='--', label='Standard Theil-Sen Slope')

# Plot the extended Theil-Sen slope (assuming intercept of mean values)
plt.plot(time, intercept + extended_slope * np.array(time), color='orange', linestyle='--', label='Extended Theil-Sen Slope')

# Adding labels and legend
plt.xlabel('Time', fontsize=20)
plt.ylabel('Temperature', fontsize=20)
# plt.title('Standard and Extended Theil-Sen Slopes with Uncertainty', fontsize=20)

plt.legend(fontsize=16)
plt.grid(True)

# Adjust tick label sizes
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

plt.show()
```
